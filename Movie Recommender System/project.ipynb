```python
# Remove unwanted columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(1)
```

```python
movies.dropna(inplace=True) # Drop null values
```

```python
import ast # To convert string to list

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
# In cast column we only want 1st 3 actors

def convert_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

```python
# In Crew column we want the director name only

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
# Change Overview column String to list of words
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# We perform an action to remove spaces between the name of words to avoid confusion during concatenation of lists
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
# Merge 'overview' 'genres'	'keywords' 'cast' 'crew' into a single column to create a 'tags' column
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
# Now we convert the list of tags to a single string because ML models work on strings not lists
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head(1)
```

```python
new_df['tags'][0]
```

```python
# Convert all text to lowercase
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(1)
```

```python
# We use stemming to reduce words to their root form so that similar words are treated the same. For example, 'running' and 'ran' will be treated as 'run'
```

```python
import nltk # Importing nltk for stemming
from nltk.stem.porter import PorterStemmer # Using PorterStemmer algorithm for stemming
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Now we do Text Vectorization where we convert text to vectors so that we can calculate similarity between vectors to find similar movies using CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english') # max_features limits the number of features to 5000 most frequent words & stop_words removes common words like is, the, a etc
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # Converting into a numpy array
```

```python
vectors
```

```python
vectors[0]
```

```python
# Now we calulate distance between each movie to all other movies using Cosine Similarity because we are dealing with high dimensional data

# In Cosine Similarity, we measure the cosine of the angle between two vectors to determine how similar they are. A cosine similarity of 1 means the vectors are identical, while a cosine similarity of 0 means they are orthogonal (completely dissimilar).
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
# Enumerate function adds a counter to an iterable and returns it in a form of enumerating object. This enumerated object can then be used directly in for loops or be converted into a list of tuples using list() method.
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11] # Get top 5 similar movies excluding itself
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
